## Entropy of the ideal gas
$S(N,T,p) = N k \Big[s_0(T_0,p_0)+ln\Big\{\big(\frac{T}{T_0}\big)^\frac{5}{2}\big(\frac{p_0}{p}\big)\Big\}\Big]$

```python
def S(N,T,p,T_0,p_0):
    return N*k*np.log((T/T_0)**(5/2)*(p_0/p))
```

```python
T_0 = 100
p_0 = 1 #show in bar
temps = np.array(range(1,500,1))
pressures = np.linspace(0.0001,1000,100)
```

```python
S_ideal_Tvary = S(6*10**23,temps,p_0,T_0,p_0)
plt.plot(temps, S_ideal_Tvary, temps,S(2*6*10**23,temps,p_0,T_0,p_0), temps,S(3*6*10**23,temps,p_0,T_0,p_0))
plt.title("Entropy for an ideal gas vs Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("Entropy")
plt.legend(["N=1 mole", "N=2 mole","N=3 mole"],loc="lower right")
plt.savefig("\image\Entropy_ideal_temp.jpg",dpi=300)
plt.show()
```

```python
S_ideal_pvary = S(6*10**23,T_0,pressures,T_0,p_0)
plt.plot(pressures, S_ideal_pvary,pressures,S(2*6*10**23,T_0,pressures,T_0,p_0),pressures,S(3*6*10**23,T_0,pressures,T_0,p_0))
plt.title("Entropy for an ideal gas vs Pressure")
plt.xlabel("Pressure (bar)")
plt.ylabel("Entropy")
plt.legend(["N=1 mole", "N=10 mole","N=50 mole"],loc="lower right")
plt.savefig("\image\Entropy_ideal_temp.jpg",dpi=300)
plt.show()
```

## Entropy of a van der Waals gas
$S(V,T) = S_0(V_0,T_0) +C_V ln\frac{T}{T_0}+N k ln\frac{V-N b}{V_0-Nb}$ 

Temperatures are around 100 K so that $C_V$ is constant. Using SI units where necessary.

```python
#constants
k = 1.380649*10**-23 #boltmann constant
R = 8.314 #gas constant
```

```python
#vdw constants
b_ideal = 0
b_ne = 0.00001709
b_he = 0.0000290
b_h20 = 0.0000305
b_o2 = 0.00003186
b_co2 = 0.0000429
b_ethanol = 0.0000871
b_butane = 0.0001164
b_dodecane = 0.0003758
```

Calculate heat $C_V$

Assuming ideal gases

```python
C_V_monoatomic = 3/2*R
C_V_diatomic = 5/2*R
C_V_polyatomic = 6/2*R
N = 10000 #mol
```

```python
def S_2(C_V,b,V,T,V_0,T_0):
    return C_V*np.log(T/T_0) + R*np.log((V-N*b)/(V_0-N*b))
```

```python
T_0 = 100
V_0 = 1 #m^3
temps = np.array(range(100,500,1))
volumes = np.linspace(0.01,100,100)
```

```python
N*b_ne
```

```python
S_vdw_vvary_ne = S_2(C_V_monoatomic,b_ne,volumes,T_0,V_0,T_0)
S_vdw_vvary_o2 = S_2(C_V_diatomic,b_o2,volumes,T_0,V_0,T_0)
S_vdw_vvary_co2 = S_2(C_V_polyatomic,b_co2,volumes,T_0,V_0,T_0)
plt.plot(volumes, S_vdw_vvary_ne,volumes, S_vdw_vvary_o2,volumes, S_vdw_vvary_co2)
plt.title("Entropy for a van der Waals gas vs Volume")
plt.xlabel("Volume ($m^{-3}$)")
plt.ylabel("Entropy")
plt.legend(["N=1 mole", "N=10 mole","N=50 mole"],loc="lower right")
plt.savefig("\image\Entropy_ideal_temp.jpg",dpi=300)
plt.show()
```

```python
S_vdw_tvary_ne = S_2(C_V_monoatomic,b_ne,V_0,temps,V_0,T_0)
S_vdw_tvary_o2 = S_2(C_V_diatomic,b_o2,V_0,temps,V_0,T_0)
S_vdw_tvary_co2 = S_2(C_V_polyatomic,b_co2,V_0,temps,V_0,T_0)
plt.plot(temps, S_vdw_tvary_ne,temps, S_vdw_tvary_o2,temps, S_vdw_tvary_co2)
plt.title("Entropy for a van der Waals gas vs Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("Entropy")
plt.legend(["N=1 mole", "N=10 mole","N=50 mole"],loc="lower right")
plt.savefig("\image\Entropy_vdw_temp.jpg",dpi=300)
plt.show()
```
